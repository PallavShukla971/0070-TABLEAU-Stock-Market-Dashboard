```python
dfs = [ apple, facebook, google, nvidia, tesla, twitter ]
# this is for moving in the data frame itself by applying a for loop
```

```python
for df in dfs:
    df["MA50"] = df.Close.rolling(50).mean
    df["MA200"] = df.Close.rolling(200).mean
```

```python
apple.head(200)
```

# Previous day close price

```python
# Daily change regarding the close price
for df in dfs:
    df['Previous Day Close Price'] = df.Close.shift(1)
```

```python
apple.head()
```

# Change In Price

```python
for df in dfs:
    df ['Change In Price'] = df['Close'] - df['Previous Day Close Price']
```

```python
apple.head()
```
